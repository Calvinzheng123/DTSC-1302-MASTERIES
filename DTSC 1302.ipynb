```python
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
import matplotlib.offsetbox as offsetbox
from matplotlib.offsetbox import OffsetImage, AnnotationBbox
import numpy as np
```

```python
df = pd.read_csv('/Users/calvi/Downloads/Average Attendance League NBA.csv')
```

```python
df.head()
```

```python
# Create a dual-axis line chart using Matplotlib
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plot population on the first y-axis
ax1.plot(df['Year'], df['Population'], color='blue', marker='o', label='Population')
ax1.set_xlabel('Year')
ax1.set_ylabel('Population', color='blue')
ax1.tick_params('y', colors='blue')

ax2 = ax1.twinx()
ax2.plot(df['Year'], df['Average Attendance'], color='green', marker='s', label='Sports Attendance')
ax2.set_ylabel('Sports Attendance', color='green')
ax2.tick_params('y', colors='green')
ax2.set_ylim((15000, 20000))

logo_path = '/Users/calvi/Downloads/Charlotte.jpeg'  # Replace with the path to your logo image
logo = plt.imread(logo_path)
imagebox = OffsetImage(logo, zoom=.5, alpha=0.1)
ab = AnnotationBbox(imagebox, (0.5, 0.5), xycoords='axes fraction', boxcoords="axes fraction", frameon=False)
ax1.add_artist(ab)

lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax2.legend(lines + lines2, labels + labels2, loc='upper left')

plt.title('Dual-Axis Line Chart: Charlotte Population and Sports Attendance Over Years')
plt.show()
```

```python
filtered_df1 = pd.read_csv('/Users/calvi/Downloads/Calvin.csv')
```

```python
filtered_df1.set_index('Variables', inplace=True)
seasons = np.arange(len(filtered_df1))

plt.figure(figsize=(12, 8))
bar_width = 0.2

for i, col in enumerate(filtered_df1.columns):
    plt.bar(seasons + i * bar_width, filtered_df1[col], bar_width, label=col)

plt.xlabel('Different Sports')
plt.ylabel('Daily Net Impact')
plt.title('Day of Event Net Impact on Average Daily Rates ($US) For Different Sports')
plt.xticks(seasons + (len(filtered_df1.columns) - 1) * bar_width / 2, filtered_df1.index)
plt.legend()
plt.show()
```
